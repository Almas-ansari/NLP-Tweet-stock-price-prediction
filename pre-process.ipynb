```python
import pandas as pd
import re
import string
```

```python
data=pd.read_csv('tesla.csv',index_col=0)
```

```python
data.shape
```

Removed unwanted columns

```python
data.drop(['id','conversation_id','created_at','timezone','link','urls','photos','thumbnail','near','geo','source','user_rt_id','user_rt','retweet_id','reply_to'],axis=1,inplace=True)
```

Only keeping tweets in english Language

```python
data=data[data.language=='en']
data.reset_index(drop=True)
```

```python
#check Url function
def check_url(tweet):
    re_equ = r"(?i)\b((?:https?://|www\d{0,3}[.]|[a-z0-9.\-]+[.][a-z]{2,4}/)(?:[^\s()<>]+|\(([^\s()<>]+|(\([^\s()<>]+\)))*\))+(?:\(([^\s()<>]+|(\([^\s()<>]+\)))*\)|[^\s`!()\[\]{};:'\".,<>?¬´¬ª‚Äú‚Äù‚Äò‚Äô]))"
    flag = bool(re.findall(re_equ, tweet))
    return flag


#remove @ and # function
def strip_all_entities(text):
    entity_prefixes = ['@','#']
    for separator in  string.punctuation:
        if separator not in entity_prefixes :
            text = text.replace(separator,' ')
    words = []
    for word in text.split():
        word = word.strip()
        if word:
            if word[0] not in entity_prefixes:
                words.append(word)
    return ' '.join(words)
```

```python
#removing tweets with "??" and links/url in it
indexes_to_be_dropped=[]
for index,row in data.iterrows():
    tweet=row['tweet']
    flag=check_url(tweet)
    if flag:
        indexes_to_be_dropped.append(index)
    if "??" in tweet:
        indexes_to_be_dropped.append(index)
        
```

```python
data=data.drop(indexes_to_be_dropped)
data.reset_index(drop=True,inplace=True)
```
